## Sigmoid Function

## $$ f(x) = \frac{1}{1 + e^{-x}} $$

## Sigmoid Curve With Python Code

```python
def sigmoid(x):
    return 1/ (1 + np.exp(-x))

x_l=np.linspace(-10,10,900)

fig = plt.figure(figsize = (12, 7))

plt.plot(x_l,sigmoid(x_l),'b')
plt.title("Sigmoid Curve", fontsize = 15)
plt.show()
```

## How do we move From a Linear Line To A Sigmoid Curve?


 ## The Math Behind
 
 
## $$ f(x) = \frac{1}{1 + e^{-(mx + b)}} $$



## $$ f(x) = \frac{1}{1 + e^{-(b_1x_1 + b_2x_2 + … + b_nx_n )}} $$

## Differences Between Linear And Logistic Regression


### Linear Regression
> * deals with continues data
> * Linear Regression is used for solving Regression problem
> * Least square estimation method is used for estimation of accuracy of the model


### Logistic Regression
> * Deals with a discrete values (yes or no)
> * Logistic regression is used for solving Classification problems
> * Maximum likelihood estimation method is used for estimation of accuracy of the model

```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
```

```python
df = pd.read_csv("../datasets/bankLoan.csv")
df.head(10)
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[["Income"]], df["Eligible"], train_size = 0.8)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
regmodel = LogisticRegression()
```

```python
regmodel.fit(X_train, y_train)
```

```python
regmodel.score(X_test, y_test)
```

```python
predictions = regmodel.predict(X_test)
```

```python
X_test
```

```python
predictions
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn import metrics
```

```python
cm = metrics.confusion_matrix(y_test, predictions)
```

```python
cm
```

```python
plt.figure(figsize = (12, 7))
sns.heatmap(cm, annot =True)
plt.xlabel("Predictions")
plt.ylabel("Truth")
plt.show()
```

```python
precision = metrics.precision_score(y_test, predictions)
precision
```
